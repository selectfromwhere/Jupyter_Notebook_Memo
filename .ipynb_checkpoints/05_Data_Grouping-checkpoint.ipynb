# Lec42_データをまとめる

```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
dframe = DataFrame({'k1':['X','X','Y','Y','Z'],
                    'k2':['alpha','beta','alpha','beta','alpha'],
                    'dataset1':np.random.randn(5),
                    'dataset2':np.random.randn(5)})
dframe
```

```python
# k1をキーにdataset1をgroupbyでまとめる
group1 = dframe['dataset1'].groupby(dframe['k1'])
group1
```

```python
# groupごとの平均値を計算
group1.mean()
```

```python
# キー項目を追加してdataset2をgroupbyする
airport = np.array(['FUK','HND','NRT','HND','FUK'])
airline = np.array(['ANA','ANA','JAL','SKY','SFJ'])
group2 = dframe['dataset2'].groupby([airport,airline])
group2
```

```python
# groupごとの平均値を出力
group2.mean()
```

```python
dframe
```

```python
# DataFrame全体をキー項目でgroupbyする
dframe.groupby('k1').mean()
```

```python
# 複合キーの設定も可能
dframe.groupby(['k1','k2']).mean()
```

```python
# dataset1に絞って出力する
dframe1 = dframe.groupby(['k1','k2'])[['dataset1']]
dframe1.mean()
```

```python
# グループごとの要素数を出力
dframe.groupby(['k1']).size()
```

```python
# 繰り返し処理によるグループごとのデータ出力
for name, group in dframe.groupby('k1'):
    print('This is the {} group'.format(name))
    print(group)
    print('¥n')
```

```python
# 複数キーの場合
for (k1,k2), group in dframe.groupby(['k1','k2']):
    print('Key1 = {} Key2 = {}'.format(k1,k2))
    print(group)
    print('¥n')
```

```python
# get_groupメソッドでデータを参照する
gr = dframe.groupby('k1')
gr.get_group('X')
```

```python
dframe
```

```python
# Listを作る
group_list = list(dframe.groupby('k1'))
group_list
```

```python
# 辞書にする
group_dict = dict(group_list)
group_dict['X']
```

```python
# 列方向で辞書を作る
group_dict_axis = dict(list(dframe.groupby(dframe.dtypes,axis=1)))
group_dict_axis
```

# Lec43_データをまとめる２

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
oxox = DataFrame(np.arange(16).reshape(4,4),
                 columns=['W','X','Y','Z'],
                 index=['ox','ox','oxo','oxox'])
oxox
```

```python
oxox.ix[1:2,['W','Y']] = np.nan
oxox
```

```python
# 列にfeelingを辞書型でマッピング
feel_map = {'W':'good','X':'bad','Y':'bad','Z':'good'}
feel_map
```

```python
# feeling[辞書型]でグループ化
oxox_col = oxox.groupby(feel_map, axis=1)
oxox_col.sum()
```

```python
# Seriesを作成する
feel_ser = Series(feel_map)
feel_ser
```

```python
# feeling[Series]でグループ化
oxox.groupby(feel_ser, axis=1).count()
```

```python
# 関数を使ってグループ化する
oxox.groupby(len).sum()
```

```python
# キー項目を追加し、関数と組み合わせてグループ化する
keys = ['A','B','A','B']
oxox.groupby([len,keys]).max()
```

# Lec44_Aggregation

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
# winequality-red.csvのダウンロード
import webbrowser
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/'
webbrowser.open(url)
```

```python
# ダウンロードしたデータの読込み。区切り文字は';'
dframe_wine = pd.read_csv('winequality-red.csv',sep=';')
dframe_wine.head(100)
```

```python
# 平均アルコール度数を参照
dframe_wine['alcohol'].mean()
```

```python
# クオリティ毎の平均値を確認
dframe_wine.groupby('quality')[['alcohol']].mean()
```

```python
# qualityでまとめてdescribe
wino = dframe_wine.groupby('quality')
wino.describe()
```

```python
# pivot_tableでもOK
dframe_wine.pivot_table(index=['quality'])
```

```python
# 最大値と最小値の差分をとる関数を定義する
def max_to_min(arr):
    return arr.max() - arr.min()
# 実行
wino.agg(max_to_min)
```

```python
# agg関数は文字列も取れる
wino.agg('max')
```

```python
wino.agg('min')
```

```python
wino.agg('mean')
```

```python
wino.agg('std')
```

```python
dframe_wine.head()
```

```python
# 列を追加する
dframe_wine['qual/alc ratio'] = dframe_wine['quality']/dframe_wine['alcohol']
dframe_wine.head()
```

```python
# alcoholでプロットして見る
%matplotlib inline
dframe_wine.plot(kind='scatter', x='quality', y='alcohol')
```

# Lec45_Split, Apply, Combine

```python
# アルコール度数の高さ順で順位付けする関数を定義する
def ranker(df):
    df['alc_content_rank'] = np.arange(len(df)) + 1
    return df
```

```python
# alcoholで並び替え
dframe_wine.sort_values('alcohol',ascending=False,inplace=True)
# qualityでグループ化して、関数を呼び出す
dframe_wine = dframe_wine.groupby('quality').apply(ranker)

dframe_wine.sort_values(['quality','alc_content_rank'])[['quality','alc_content_rank']]
```

```python
# quality毎の本数を算出する
dframe_wine['quality'].value_counts()
```

```python
# それぞれのqualityからアルコール度数の一番高いワインを参照する
dframe_wine[dframe_wine.alc_content_rank == 1].sort_values('quality')
```

# Lec46_クロス集計表

```python
import pandas as pd
from io import StringIO
```

```python
data ="""Sample  Animal   Intelligence
1   Dog   Dumb
2 Dog Dumb
3   Cat Smart
4 Cat    Smart
5 Dog Smart
6 Cat Smart"""
dframe = pd.read_table(StringIO(data),sep='\s+')
# sep='\s+' は正規表現
dframe
```

```python
# クロス集計表を作成する。margins=Trueで合算値を出力
pd.crosstab(dframe.Animal, dframe.Intelligence, margins=True)
```

```python
pd.crosstab(dframe.Intelligence, dframe.Animal, margins=True)
```
